```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Assuming 'df' is your DataFrame
# Sort data by 'month' to maintain the temporal order
df['month'] = pd.to_datetime(df['month'])
df = df.sort_values('month')

# Extract relevant features
numeric_df = df[['floor_area_sqm', 'resale_price']].copy()

# Normalize data (LSTMs are sensitive to the scale of input data)
scaler = MinMaxScaler()
numeric_df[['floor_area_sqm', 'resale_price']] = scaler.fit_transform(numeric_df[['floor_area_sqm', 'resale_price']])

# Convert data to sequences for LSTM
def create_sequences(data, sequence_length):
    sequences = []
    for i in range(len(data) - sequence_length):
        seq = data[i:i+sequence_length]
        sequences.append(seq)
    return np.array(sequences)

# Define sequence length (adjust based on your preference)
sequence_length = 10

# Create sequences for LSTM
sequences = create_sequences(numeric_df.values, sequence_length)

# Split data into features (X) and target variable (y)
X = sequences[:, :-1, :]  # All columns except the last one
y = sequences[:, -1, 1]    # The last column (resale_price)

# Split data into training and testing sets
split_ratio = 0.8
split_index = int(split_ratio * len(sequences))

X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

# Build LSTM model
model = Sequential()
model.add(LSTM(50, input_shape=(X.shape[1], X.shape[2])))
model.add(Dense(1))  # Output layer with one neuron for regression

model.compile(optimizer='adam', loss='mse')  # Mean Squared Error as loss function

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.1)

# Make predictions on the test set
lstm_predictions = model.predict(X_test)

# Inverse transform predictions to original scale
lstm_predictions = scaler.inverse_transform(np.concatenate((X_test[:, -1, :-1], lstm_predictions.reshape(-1, 1)), axis=1))[:, -1]

# Inverse transform actual values to original scale
y_test = scaler.inverse_transform(np.concatenate((X_test[:, -1, :-1], y_test.reshape(-1, 1)), axis=1))[:, -1]

# Evaluate the model
lstm_mae = mean_absolute_error(y_test, lstm_predictions)
print(f'LSTM Mean Absolute Error: {lstm_mae}')
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error

# Assuming 'df' is your DataFrame
# Sort data by 'month' to maintain the temporal order
df['month'] = pd.to_datetime(df['month'])
df = df.sort_values('month')

# Extract relevant features
numeric_df = df[['floor_area_sqm', 'resale_price']].copy()

# Normalize data (LSTMs are sensitive to the scale of input data)
scaler = MinMaxScaler()
numeric_df[['floor_area_sqm', 'resale_price']] = scaler.fit_transform(numeric_df[['floor_area_sqm', 'resale_price']])

# Convert data to sequences for LSTM
def create_sequences(data, sequence_length):
    sequences = []
    for i in range(len(data) - sequence_length):
        seq = data[i:i+sequence_length]
        sequences.append(seq)
    return np.array(sequences)

# Define sequence length (adjust based on your preference)
sequence_length = 10

# Create sequences for LSTM
sequences = create_sequences(numeric_df.values, sequence_length)

# Split data into features (X) and target variable (y)
X = sequences[:, :-1, :]  # All columns except the last one
y = sequences[:, -1, 1]    # The last column (resale_price)

# Convert NumPy arrays to PyTorch tensors
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32)

# Split data into training and testing sets
split_ratio = 0.8
split_index = int(split_ratio * len(sequences))

X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

# Define LSTM model
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        _, (h_n, _) = self.lstm(x)
        out = self.fc(h_n[-1, :, :])
        return out

# Instantiate the model
input_size = X.shape[2]
hidden_size = 50
output_size = 1
lstm_model = LSTMModel(input_size, hidden_size, output_size)

# Define loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(lstm_model.parameters(), lr=0.001)

# Train the model
num_epochs = 50
for epoch in range(num_epochs):
    outputs = lstm_model(X_train)
    optimizer.zero_grad()
    loss = criterion(outputs.squeeze(), y_train)
    loss.backward()
    optimizer.step()

# Make predictions on the test set
lstm_predictions = lstm_model(X_test).detach().numpy()

# Inverse transform predictions to original scale
lstm_predictions = scaler.inverse_transform(np.concatenate((X_test[:, -1, :-1], lstm_predictions.reshape(-1, 1)), axis=1))[:, -1]

# Inverse transform actual values to original scale
y_test = scaler.inverse_transform(np.concatenate((X_test[:, -1, :-1], y_test.reshape(-1, 1)), axis=1))[:, -1]

# Evaluate the model
lstm_mae = mean_absolute_error(y_test, lstm_predictions)
print(f'LSTM Mean Absolute Error: {lstm_mae}')
```
